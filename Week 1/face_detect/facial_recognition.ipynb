```python
# import cv2 and sys!
import cv2
import sys
import matplotlib.pyplot as plt 
```

### Paths

Now we are going to include the paths to our photo we want to detect a face in as well as our cascade classifier file. You should not change the path to the cascade classifier file, but you can change the path to the image after going through this notebook once!


```python
imagePath = "./scientists.jpg" #change this line to do different images!
cascPath = "./haarcascade_frontalface_default.xml" #do not change this line -- it is how our face classifier works!
```

### Rest of the Tutorial

Now you should be able to go through the rest of the tutorial! Below is the copied code from the tutorial with some adjustments for CV3. Please read through this tutorial along side it to understand how the code works: [https://realpython.com/face\-recognition\-with\-python/.](https://realpython.com/face-recognition-with-python/)


```python
faceCascade = cv2.CascadeClassifier(cascPath)
```

```python
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30),
    flags = cv2.CASCADE_SCALE_IMAGE
)
```

```python
print("Found {0} faces!".format(len(faces)))

# Draw a rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
